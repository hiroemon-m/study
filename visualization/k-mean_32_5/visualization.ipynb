```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cbook as cbook
```

```python
attr_auc = []
attr_nll = []
edge_auc = []
edge_nll = []


for i in [3,4,6,8,16]:
    attr_auc_path = "/Users/matsumoto-hirotomo/study/experiment_data/hard/k-mean_32_5/persona = {}/proposed_attr_auc.npy".format(str(i))
    attr_nll_path = "/Users/matsumoto-hirotomo/study/experiment_data/hard/k-mean_32_5/persona = {}/proposed_attr_nll.npy".format(str(i))
    edge_auc_path = "/Users/matsumoto-hirotomo/study/experiment_data/hard/k-mean_32_5/persona = {}/proposed_edge_auc.npy".format(str(i))
    edge_nll_path = "/Users/matsumoto-hirotomo/study/experiment_data/hard/k-mean_32_5/persona = {}/proposed_edge_nll.npy".format(str(i))

    attr_auc.append(np.load(attr_auc_path))
    attr_nll.append(np.load(attr_nll_path))
    edge_auc.append(np.load(edge_auc_path))
    edge_nll.append(np.load(edge_nll_path))

attr_mean = np.mean(attr_auc,axis=1)
attr_max = np.max(attr_auc,axis=1)
attr_min = np.min(attr_auc,axis=1)
edge_mean = np.mean(edge_auc,axis=1)
edge_max = np.max(edge_auc,axis=1)
edge_min = np.min(edge_auc,axis=1)
```

```python

fig, ax = plt.subplots(1)
t = [1,2,3,4,5]

ax.plot(t, edge_mean[0], lw=2, label='persona3', color='blue')
ax.plot(t, edge_mean[1], lw=2, label='persona4', color='red')
ax.plot(t, edge_mean[2], lw=2, label='persona6', color='orange')
ax.plot(t, edge_mean[3], lw=2, label='persona8', color='green')
ax.plot(t, edge_mean[4], lw=2, label='persona16', color='sienna')
#ax.plot(t, edge_mean[5], lw=2, label='persona32', color='teal')
#ax.plot(t, edge_mean[6], lw=2, label='persona16', color='indigo')

ax.fill_between(t, edge_min[0], edge_max[0], facecolor='blue', alpha=0.1)
ax.fill_between(t, edge_min[1], edge_max[1], facecolor='red', alpha=0.1)
ax.fill_between(t, edge_min[2], edge_max[2], facecolor='orange', alpha=0.1)
ax.fill_between(t, edge_min[3], edge_max[3], facecolor='green', alpha=0.1)
ax.fill_between(t, edge_min[4], edge_max[4], facecolor='sienna', alpha=0.1)
#ax.fill_between(t, edge_min[5], edge_max[5], facecolor='teal', alpha=0.1)
#ax.fill_between(t, edge_min[6], edge_max[6], facecolor='indigo', alpha=0.1)

ax.set_title(r'edge auc')
ax.legend(loc='upper left')
ax.set_xlabel('num steps')
ax.set_ylabel('position')
```
